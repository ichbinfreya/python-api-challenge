## Starter Code to Generate Random Geographic Coordinates and a List of Cities

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import the OpenWeatherMap API key
from api_keys import weather_api_key as api_key

# Import citipy to determine the cities based on latitude and longitude
from citipy import citipy
```

### Generate the Cities List by Using the `citipy` Library

```python
# Empty list for holding the latitude and longitude combinations
lat_lngs = []

# Empty list for holding the cities names
cities = []

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name

    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(f"Number of cities in the list: {len(cities)}")
```

### Use the OpenWeatherMap API to retrieve weather data from the cities list generated in the started code

```python
# Set the API base URL
url = "http://api.openweathermap.org/data/2.5/weather?"

# Define an empty list to fetch the weather data for each city
city_data = []

# Print to logger
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop through all the cities in our list to fetch weather data
deno = len(cities)//40
for i, city in enumerate(cities[:deno*40]):

    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 0

    # Create endpoint URL with each city
    city_url = f"{url}q={city}&appid={api_key}&units=metric"

    # Log the url, record, and set numbers (print only the first few and then periodically)
    if i % deno == 0:  # Adjust this condition to see 40 pieces of the data
        print("Processing Record %s of Set %s | %s" % (record_count, set_count, city))

    # Add 1 to the record count
    record_count += 1

    # Run an API request for each of the cities
    try:
        # Parse the JSON and retrieve data
        city_weather = requests.get(city_url).json()

        # Parse out latitude, longitude, max temp, humidity, cloudiness, wind speed, country, and date
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_date = city_weather["dt"]

        # Append the City information into city_data list
        city_data.append({"City": city,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})

    # If an error is experienced, skip the city
    except:
        print("City not found. Skipping...")
        pass

    # pause to avoid rate limiting
    time.sleep(1)

# Indicate that Data Loading is complete
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Convert the cities weather data into a Pandas DataFrame
city_data_df = pd.DataFrame(city_data)

# Show Record Count
city_data_df.count()
```

## Requirement 2: Compute Linear Regression for Each Relationship


```python
# Create a DataFrame with the Northern Hemisphere data (Latitude >= 0)
northern_hemi_df = city_data_df[city_data_df["Lat"] >= 0]

# Display sample data
northern_hemi_df.head()
```

```python
# Create a DataFrame with the Southern Hemisphere data (Latitude < 0)
southern_hemi_df = city_data_df[city_data_df["Lat"] <= 0]

# Display sample data
southern_hemi_df.head()
```

###  Temperature vs. Latitude Linear Regression Plot

```python
# Define a function to create Linear Regression plots
def plot_linear_regression(x_values, y_values, title, xlabel, ylabel):
    
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = f"y = {slope:.2f}x + {intercept:.2f}"

    # Plot data
    plt.scatter(x_values, y_values)
    plt.plot(x_values, regress_values, "r")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.annotate(line_eq, xy=(x_values.min() + 1, y_values.min() + 1), fontsize=15, color="red")
    plt.show()

    # Print r-squared value
    print(f"The r^2-value is: {r_value**2}")
```

```python
# Linear regression on Northern Hemisphere
relationships = [
    ("Lat", "Max Temp", "Latitude vs. Max Temp", "Latitude", "Max Temp (C)")
]

for x_col, y_col, title, xlabel, ylabel in relationships:
    x_values = northern_hemi_df[x_col]
    y_values = northern_hemi_df[y_col]
    plot_linear_regression(x_values, y_values, f"Northern Hemisphere {title}", xlabel, ylabel)
```

**Discussion about the linear relationship:** This graph shows the relationship between latitude and maximum temperature in the Northern Hemisphere, which indicates that for every one-degree increase in latitude, the maximum temperature decreases by 0.22 degrees Celsius. It is an inverse relationship between latitude and maximum temperature. As an increase in latitude, which means moving nothwards, the maximum temperature generally decreases. The intercept, which is 31.83, represents the estimated maximum temperature when the latitude is 0. The equator receives more direct sunlight throughout the year. That's why the intercept is higher than that of other latitudes. In addtion, the data points are spread around the regression line, suggesting that while latitude is a significant factor, there may be other variables influencing maximum temperature. 

```python
# Linear regression on Southern Hemisphere
relationships = [("Lat", "Max Temp", "Latitude vs. Max Temp", "Latitude", "Max Temp (C)")]

for x_col, y_col, title, xlabel, ylabel in relationships:
    x_values = southern_hemi_df[x_col]
    y_values = southern_hemi_df[y_col]
    plot_linear_regression(x_values, y_values, f"Southern Hemisphere {title}", xlabel, ylabel)
```

**Discussion about the linear relationship:** This graph shows the relationship between latitude and maximum temperature in the Southern Hemisphere. Comparing to the Northern Hemisphere, the Southern Hemisphere's latitude increases as it moves closer to the equator from the South Pole.The equation indicates that for every one-degree increase in latitude, the maximum temperature increases by 0.52 degrees Celsius. This is a direct relationship between latitude and maximum temperature in the Southern Hemisphere. As you move closer to the equator, the maximum temperature increases. The intercept, which is 29.95, represents the estimated maximum temperature at the equator (latitude 0). This is consistent with the high temperatures typically observed at the equator. Similar to the Northern Hemisphere, there is a spread of data points around the regression line, indicating other influencing factors on maximum temperature beyond latitude.

**Discussion about comparing Northern and Soutern linear relationship:** 
Please note that each run will be generated different data. The numbers below is based on the current run.
Firstly, latitude can be a predictor of maximum temperature based on two graphs presentation, with temperatures generally decreasing as one moves away from the equator.
Secondly, the steeper slope (0.52) in the Southern Hemisphere compared to the Northern Hemisphere (-0.22) suggests that temperature changes more rapidly with latitude in the Southern Hemisphere. This means the Southern Hemisphere shows a stronger relationship between latitude and maximum temperature compared to the Northern Hemisphere. However, various geographical and climatic factors specific to the Southern Hemisphere may reult in the difference between two steeper slopes. This might be due to the larger proportion of ocean in the Southern Hemisphere, which can influence temperature distribution.
Finally, the spread of data points in both plots indicates that while latitude is an important factor, other variables such as altitude, ocean proportions (which can have impact on water vaporation and amount of rainfalls), and prevailing winds also play crucial roles in determining maximum temperature.
Overall, these linear regressions help visualize the expected trend of decreasing temperature with increasing distance from the equator. However, we need to address the possibility of multiple factors derived from complex climatic systems to determine local temperatures.
